```python
import pandas as pd

x1 = []
x2 = []
x3 = []
x4 = []
x5 = []

with open("E:\data_banknote_authentication.txt") as file:
    for line in file:
        line = line.split("\n")[0]
        list = line.split(",") 
        x1.append(list[0])
        x2.append(list[1])
        x3.append(list[2])
        x4.append(list[3])
        x5.append(list[4])

table = pd.DataFrame({})
for i in range(len(x1)):
    table[i+1] = [x1[i], x2[i], x3[i], x4[i], x5[i]]
table.index = ["x1", "x2", "x3", "x4", "x5"]
table.head()
```

```python
import random

train_x = []
for i in range(60):
    index = random.randint(0, 22) + i*23;
    if index < len(x1):
        train_x.append([float(x1[index]), float(x2[index]), float(x3[index]), float(x4[index]), float(x5[index])])
    else: 
        train_x.append([float(x1[len(x1)-1]), float(x2[len(x1)-1]), float(x3[len(x1)-1]), float(x4[len(x1)-1]), float(x5[len(x1)-1])])
        
table = pd.DataFrame({})
for i in range(len(train_x)):
    table[i+1] = [train_x[i][0], train_x[i][1], train_x[i][2], train_x[i][3], train_x[i][4]]
table.index = ["x1", "x2", "x3", "x4", "x5"]
table.head()
```

```python
w = []
for i in range(10):
    l = []
    for j in range(5):
        l.append(random.uniform(-1, 1))
    w.append(l)
    
table = pd.DataFrame({})
for i in range(len(w)):
    table[i+1] = [w[i][0], w[i][1], w[i][2], w[i][3], w[i][4]]
table.index = ["x1", "x2", "x3", "x4", "x5"]
table.head()
```

```python
import math

for i in range(60):
    d_min = math.sqrt((train_x[i][0]-w[0][0])**2 + (train_x[i][1]-w[0][1])**2 + (train_x[i][2]-w[0][2])**2 + (train_x[i][3]-w[0][3])**2+ (train_x[i][4]-w[0][4])**2)
    index = 0
    for j in range(1, 10):
        d = math.sqrt((train_x[i][0]-w[j][0])**2 + (train_x[i][1]-w[j][1])**2 + (train_x[i][2]-w[j][2])**2 + (train_x[i][3]-w[j][3])**2+ (train_x[i][4]-w[j][4])**2)
        if d < d_min:
            d_min = d
            index = j
    for j in range(5):
        w[index][j] += 0.04*(train_x[i][j] - w[index][j])
        
table = pd.DataFrame({})
for i in range(len(w)):
    table[i+1] = [x1[i], x2[i], x3[i], x4[i], x5[i]]
table.index = ["x1", "x2", "x3", "x4", "x5"]
table.head()
```

```python
cluster = []
x = []
for i in range(len(x1)):
    x.append([float(x1[i]), float(x2[i]), float(x3[i]), float(x4[i]), float(x5[i])])
    
for i in range(len(x)):
    d_min = math.sqrt((x[i][0]-w[0][0])**2 + (x[i][1]-w[0][1])**2 + (x[i][2]-w[0][2])**2 + (x[i][3]-w[0][3])**2+ (x[i][4]-w[0][4])**2)
    index = 0
    for j in range(1, 10):
        d = math.sqrt((x[i][0]-w[j][0])**2 + (x[i][1]-w[j][1])**2 + (x[i][2]-w[j][2])**2 + (x[i][3]-w[j][3])**2+ (x[i][4]-w[j][4])**2)
        if d < d_min:
            d_min = d
            index = j
    cluster.append(index+1) 
    
table = pd.DataFrame({})
for i in range(len(cluster)):
    table[i+1] = [x1[i], x2[i], x3[i], x4[i], x5[i], cluster[i]]
table.index = ["x1", "x2", "x3", "x4", "x5", "cluster"]
table
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

z = []
for i in range(5):
    l = []
    for j in range(len(x)):
        l.append(x[j][i])
    z.append(l)
    
s = [z[4][i]*10+10 for i in range(len(x5))]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(z[0], z[1], z[2], label='text', s=s, c=cluster)
```
